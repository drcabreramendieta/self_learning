```python
import scipy.io
import numpy as np
import torch
import matplotlib.pyplot as plt
plt.ion()
```

## Original signal analysis

```python
'''original signal'''

mat = scipy.io.loadmat('Ep_019_resample_new.mat')
x_1 = mat['train_data']
fs = 50000
N = 4096
t = np.arange(N)/fs
x_1 = torch.from_numpy(x_1)
x_1 = x_1.float()
x_1 = x_1[:,None,:]
plt.plot(t, x_1[0][0])
```

Spectrum:

```python
'''original fft'''

YFFT = torch.fft.fft(x_1)
YFFT = YFFT/N
mag = torch.abs(YFFT)
angle = torch.angle(YFFT)

F_coef = torch.fft.fftfreq(N, 1/fs)
plt.subplot(211)
plt.plot(F_coef, mag[0][0])
plt.xlim([-100,100])
plt.subplot(212)
plt.plot(F_coef, angle[0][0])
plt.xlim([-100,100])
```

## Downsampling

```python
'''downsampling'''
scale_factor = 0.9
d_N = int(N * scale_factor)
d_s = torch.nn.functional.interpolate(x_1, scale_factor=scale_factor, mode='nearest')
plt.plot(d_s[0][0])
d_YFFT = torch.fft.fft(d_s)
d_YFFT = d_YFFT/int(N*scale_factor)
d_mag = torch.abs(d_YFFT)
d_angle = torch.angle(d_YFFT)

d_F_coef = torch.fft.fftfreq(len(d_YFFT[0][0]), d=1/(fs*0.9))
```

## Upsampling with shift

```python
d_YFFT_shift = torch.fft.fftshift(d_YFFT)
d_YFFT_shift = torch.concat((d_YFFT_shift,d_YFFT_shift[:,:,0][:,:,None]),dim=2)
d_YFFT_real_up = torch.nn.functional.interpolate(d_YFFT_shift.real, size=N+1, mode='linear')
d_YFFT_im_up = torch.nn.functional.interpolate(d_YFFT_shift.imag, size=N+1, mode='linear')
d_YFFT_up_shift = d_YFFT_real_up + (1j)*d_YFFT_im_up

d_YFFT_up_shift = d_YFFT_up_shift[:,:,:-1]
d_YFFT_up = torch.fft.ifftshift(d_YFFT_up_shift)
d_YFFT_up_mag = torch.abs(d_YFFT_up)
d_YFFT_up_angle = torch.angle(d_YFFT_up)
plt.subplot(211)
plt.plot(F_coef, d_YFFT_up_mag[0][0])
plt.xlim([-700,700])
plt.subplot(212)
plt.plot(F_coef, d_YFFT_up_angle[0][0])
plt.xlim([-300,300])

plt.figure()
u_up = torch.fft.ifft(N*d_YFFT_up)
plt.plot(t,u_up[0][0])
plt.show()
```

## Upsampling without shift

```python
d_YFFT_real_up = torch.nn.functional.interpolate(d_YFFT.real, size=N, mode='linear')
d_YFFT_im_up = torch.nn.functional.interpolate(d_YFFT.imag, size=N, mode='linear')
d_YFFT_up = d_YFFT_real_up + (1j)*d_YFFT_im_up
d_YFFT_up_mag = torch.abs(d_YFFT_up)
d_YFFT_up_angle = torch.angle(d_YFFT_up)
plt.subplot(211)
plt.plot(F_coef, d_YFFT_up_mag[0][0])
plt.xlim([-100,100])
plt.subplot(212)
plt.plot(F_coef, d_YFFT_up_angle[0][0])
plt.xlim([-100,100])

plt.figure()
u_up = torch.fft.ifft(N*d_YFFT_up)
plt.plot(t,u_up[0][0])
plt.show()
```
